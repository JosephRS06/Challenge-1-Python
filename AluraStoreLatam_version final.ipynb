#1. Análisis de facturación



```python
def ingresoTotal(tienda):
  '''
  Funcion para sumar el ingreso total de cada tienda
  '''
  sumaIngresos = tienda["Precio"].sum()
  print(f'El ingreso total de la tienda es de: {sumaIngresos:,.2f}')
  return sumaIngresos
```

```python
ingresoTotal(tienda)
ingresoTotal(tienda2)
ingresoTotal(tienda3)
ingresoTotal(tienda4)
```

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = []
# Recorremos las tiendas y calculamos el ingreso total para cada una
for i, tienda in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    suma = ingresoTotal(tienda)  # Llamamos a la función ingresoTotal() para cada tienda
    ingresos.append(suma)  # Almacenamos el ingreso total de cada tienda

# Graficar los ingresos totales en un gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color='skyblue')  # Barras de color celeste
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total ($)')
plt.xticks(rotation=45)  # Rotar las etiquetas de las tiendas para mejor legibilidad
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Cuadrícula en el eje Y
plt.tight_layout()  # Ajustar el diseño
plt.show()
```

# 2. Ventas por categoría

```python
'''
Para contar las ventas de cada categoria, si solo quisieramos mostrar el top 3 ventas de cada tienda se puede utilizar .head(3)
ventasCategoria = tienda ['Categoría del Producto'].value_counts().head(3)
'''
```

```python
def ventas(tienda):
  '''
  Funcion para analizar las ventas por categoria de cada tienda
  '''
  VentasCategoria = tienda['Categoría del Producto'].value_counts()
  return VentasCategoria
```

```python
ventas(tienda4)
```

```python
for i, tienda in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    ventas_categoria = ventas(tienda)  # Obtener las ventas por categoría
    plt.figure(figsize=(7, 7))  # Cuadrado para que el círculo quede bien
    plt.pie(
        ventas_categoria.values,  # Los tamaños de las porciones
        labels=ventas_categoria.index,  # Las etiquetas de las categorías
        autopct='%1.1f%%',  # Para mostrar porcentaje con 1 decimal
        startangle=140,  # Rotación inicial para mejor apariencia
    )
    plt.title(f'Porcentaje de Ventas por Categoría - Tienda {i}')
    plt.axis('equal')  # Para que el círculo no quede ovalado
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
'''
tambien se puede utilizar la funcion .mean() para obtener el promedio de la columna
calificacionPrmedio = tienda['Calificación'].mean()
'''
```

```python
def calificacionPromedio(tienda):
  '''
  Funcion para analizar la calificacion promedio de cada tienda, con el tres decimales => round(numero, 3)
  '''
  promedioCalificacion = round(tienda['Calificación'].sum()/len(tienda['Calificación']), 3)
  print(f'El promedio de clasificacion segun los clientes de la tienda es el siguientes: {promedioCalificacion}')
  return promedioCalificacion
```

```python
calificaciones_promedio = []
# Recorremos las tiendas y calculamos la calificación promedio para cada una
for i, tienda in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    calificacion = calificacionPromedio(tienda)  # Llamamos a la función calificacionPromedio()
    calificaciones_promedio.append(calificacion)  # Almacenamos la calificación promedio de cada tienda
```

```python
plt.figure(figsize=(8, 5))
plt.plot(tiendas, calificaciones_promedio, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)  # Línea azul con círculos
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Calificación Promedio')
plt.xticks(rotation=45)  # Rotar las etiquetas de las tiendas para mejor legibilidad
plt.grid(True, linestyle='--', alpha=0.7)  # Cuadrícula
plt.tight_layout()  # Ajustar el diseño
plt.show() # mostramos el grafico
```

# 4. Productos más y menos vendidos

```python
def productosVendidos(tienda):
  '''
  Para analizar los productos de mas a menos vendidos
  '''
  productosVendidos = tienda['Producto'].value_counts()
  for producto, cantidad in productosVendidos.items():
    print(f'{producto}: {cantidad}')
    return productosVendidos
```

```python
from folium.plugins import HeatMap
```

```python
def coordenadas(tienda):
  '''
  Funcion para obtener las coordenadas de cada tienda
  '''
  latitudes = tienda['lat'].to_numpy()  # Convierte la columna 'lat' a un array de numpy
  longitudes = tienda['lon'].to_numpy()  # Convierte la columna 'lon' a un array de numpy
  coordenadas_lista = list(zip(latitudes, longitudes))  # Combina latitudes y longitudes en tuplas
  return coordenadas_lista  # Devuelve la lista de tuplas
```

```python
coordenadas(tienda)
```

```python
def crear_heatmap(coordenadas_lista):
    # Crear el mapa base centrado en el promedio de las coordenadas
    mapa = folium.Map(location=[sum(lat for lat, lon in coordenadas_lista) / len(coordenadas_lista),
                                sum(lon for lat, lon in coordenadas_lista) / len(coordenadas_lista)], zoom_start=12)
    HeatMap(coordenadas_lista).add_to(mapa) # Crear el Heatmap
    return mapa
coordenadas_lista = coordenadas(tienda)  # se llama a la función para obtener las coordenadas
mapa = crear_heatmap(coordenadas_lista)  # se pasa la lista de coordenadas
```
